```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel file into a Pandas dataframe
df = pd.read_excel('encrypted_images.xlsx')
df = df.sort_values(by=['Image_Size'])
ax = df.plot.bar(x='Image_Size', y='Encryption_Time')

# Remove x-axis tick labels
ax.set_xticklabels([])

# Set x and y axis labels and title
plt.xlabel('Image Size')
plt.ylabel('Encryption Time (seconds)')
plt.title('Encryption Time vs Image Size')

# Show the graph
plt.show()
```

```python
x = df['Image_Size']
y = df['Encryption_Time']

# Plot the data as a scatter plot
plt.plot(x, y, 'o', label='Data')
plt.title('Image_Size vs Encryption_Time')
plt.xlabel('Image_Size')
plt.ylabel('Encryption_Time')
plt.legend()
plt.show()
```

```python
correlation = df['Image_Size'].corr(df['Encryption_Time'])
print(f"Correlation between Image_Size and Encryption_Time: {correlation}")
```
